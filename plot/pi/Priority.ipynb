```python
import json
import matplotlib.pyplot as plt
import numpy as np
rtime = []
prio = []
f = open("ertekek.json")
data = json.load(f)
for i in data["measurements"]:
    prio.append(i["parameters"]["prio"])
    min = i["results"][0]
    for j in i["results"]:
        if (j<min):
            min = j
    rtime.append(min)
plt.plot(prio,rtime,"o")
plt.show()
#Itt az öt futási mintából a legjobbat választottam ki, x-priority(mozgó paraméter, többi rögzített) y-runtime
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np
rtime = []
min_gran = []
first=0
second=0
third=0
forth=0
f = open("ertekek.json")
data = json.load(f)
for i in data["measurements"]:
    count=0
    for j in i["results"]:
        if(j<3.6):
            count+=1
    if(i["parameters"]["prio"]=="-20"):
        first+=count
    elif(i["parameters"]["prio"]=="-6"):
        second+=count
    elif(i["parameters"]["prio"]=="6"):
        third+=count
    elif(i["parameters"]["prio"]=="19"):
        forth+=count
print("\n","-20: \t",first,"db","\n"," -6: \t",second,"db","\n","  6: \t",third,"db","\n"," 19: \t",forth,"db")
mylabels=np.array(["-20","-6","6","19"])
y = np.array([first,second,third,forth])
plt.bar(mylabels,y)
plt.show()
#Itt az figyelhető meg, hogy adott beállításon, mennyi olyan minta volt amely 3.6 másodperc alatt végzett.
```
